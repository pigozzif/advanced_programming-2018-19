```python
#Lambda

def reverse_order(sequence):
    '''
    Given a list of words, construct a new one sorted in reversed
    spelling order, using a lambda function
    '''
    return sorted(sequence, key=lambda x: x[::-1])
```

```python
#Sequence

class Sequence(object):
    
    def __init__(self, iterable):
        '''
        Constructor builds an internal list from an iterable
        '''
        from copy import deepcopy
        self.s = deepcopy(list(iterable))
        
    def __str__(self):
        '''
        For printing the elements of the internal list
        '''
        return str(self.s)
    
    def __len__(self):
        '''
        Returns the length of the internal list
        '''
        return len(self.s)
    
    def __eq__(self, other_s):
        '''
        Tests if the internal lists are the same. Returns a boolean
        '''
        return False if len(self.s) != len(other_s) else all([self[num] == other_s[num] for num in range(len(other_s))])
    
    def __getitem__(self, key):
        '''
        Returns element in position key of the internal list
        '''
        return self.s[key]
    
    def __setitem__(self, key, item):
        '''
        Overwrites element in position key of the internal list with item
        '''
        self.s[key] = item
        
    def __delitem__(self, key):
        '''
        Deletes the slice in the internal list corresponding to key
        '''
        del self.s[key]
        
    def __iter__(self):
        '''
        Returns an iterable of the class
        '''
        for elem in self.s:
            yield elem
            
    def __copy__(self):
        '''
        Returns a deepcopy
        '''
        from copy import deepcopy
        return Sequence(deepcopy(self))
            
    def insert(self, position, value):
        '''
        Inserts value at position of the internal list
        '''
        self.s[position] = value
        
    def append(self, value):
        '''
        Appends value at the end of the internal list
        '''
        self.s.append(value)     
```
